```python
import pandas as pd
import ast

# === 1. Load all datasets ===
# Load groundwater_df by skipping the header rows and manually setting column names
groundwater_df = pd.read_csv("/content/drive/MyDrive/GW_Levels_1750139546761.csv", skiprows=[0, 1])
# Manually assign column names based on inspection (11 columns after skipping header)
groundwater_df.columns = ["District", "4/25/01", "5/24/01", "11/24/01", "5/25/01", "Rise_4/25/01", "Rise_5/24/01", "Fall_11/24/01", "Rainfall_Normal", "Rainfall_Actual", "Rainfall_Deviation"] # Corrected column names
# Select District and a suitable Groundwater column (based on previous display output, assuming Rainfall_Actual is the relevant one after skipping)
groundwater_df = groundwater_df[["District", "Rainfall_Actual"]].dropna() # Selected Rainfall_Actual as a proxy for Groundwater based on header
groundwater_df = groundwater_df.rename(columns={"Rainfall_Actual": "Groundwater"}) # Renamed to Groundwater
groundwater_df["Groundwater"] = pd.to_numeric(groundwater_df["Groundwater"], errors='coerce')
groundwater_df = groundwater_df.dropna(subset=["Groundwater"])


# Load rainfall_df by skipping the header rows and manually setting column names
rainfall_df = pd.read_csv("/content/drive/MyDrive/Rainfall_Summary_1750141935618.csv", skiprows=[0, 1])
# Manually assign column names based on inspection (16 columns after skipping header)
rainfall_df.columns = ["District", "4/25/01_Normal", "4/25/01_Actual", "5/25/01_Normal", "5/25/01_Actual", "Yesterday_Normal", "Yesterday_Actual", "Jan-Jun_Normal", "Jan-Jun_Actual", "Jun-Jun_Normal", "Jun-Jun_Actual", "Rainy_Days", "Jun-Jun_Total_Normal", "Jun-Jun_Total_Actual", "Deviation", "Status"] # Assuming 16 columns
# Select District and Rainfall columns and convert Rainfall to numeric
rainfall_df = rainfall_df[["District", "Jun-Jun_Total_Actual"]].dropna()
rainfall_df = rainfall_df.rename(columns={"Jun-Jun_Total_Actual": "Rainfall"})
rainfall_df["Rainfall"] = pd.to_numeric(rainfall_df["Rainfall"], errors='coerce')
rainfall_df = rainfall_df.dropna(subset=["Rainfall"])


# Load soil_df (Soil suitability)
soil_df = pd.read_csv("/content/drive/MyDrive/Soil_Crop_Suitability_AP.csv")
soil_df.columns = soil_df.columns.str.strip()
soil_df = soil_df.rename(columns={"Soil Type": "Soil_Type"})
soil_df = soil_df.dropna(subset=["Soil_Type"]) # Soil_df does not have a District column, so we drop rows with missing soil type


# Load crop_df (rules)
crop_df = pd.read_csv("/content/drive/MyDrive/AP_Crop_Dataset.csv")
crop_df.columns = crop_df.columns.str.strip()

# Safely evaluate season, soil, and districts columns
def safe_literal_eval(x):
    try:
        return ast.literal_eval(x) if isinstance(x, str) else x
    except (ValueError, SyntaxError):
        return x # Return original value if evaluation fails

if 'season' in crop_df.columns:
  crop_df["season"] = crop_df["season"].apply(safe_literal_eval)
if 'soil' in crop_df.columns:
  crop_df["soil"] = crop_df["soil"].apply(safe_literal_eval)
if 'districts' in crop_df.columns:
  crop_df["districts"] = crop_df["districts"].apply(safe_literal_eval)


# Load mixed_df
mixed_df = pd.read_csv("/content/drive/MyDrive/district_mixed_cropping_suggestions.csv")
mixed_df.columns = mixed_df.columns.str.strip()


# === 6. Merge environmental datasets ===
merged_df = pd.merge(rainfall_df, groundwater_df, on="District", how="inner")
# We cannot merge soil_df directly as it does not have a district column,
# so we will use it separately in the recommendation function


# === 7. Define Rule-Based Crop Recommender ===
def recommend_crops(district_name, season, temperature, soil_type):
    try:
        # Find the environmental data for the given district
        env_data = merged_df[merged_df["District"] == district_name]
        if env_data.empty:
            return [{"crop": f"Error: No environmental data found for {district_name}", "suitable_for_mixed_cropping": False}]

        rainfall = env_data.iloc[0]["Rainfall"]
        groundwater = env_data.iloc[0]["Groundwater"]

        # Find suitable soil types from the soil_df
        suitable_soil_row = soil_df[soil_df["Soil_Type"].str.contains(soil_type, case=False, na=False)]
        if suitable_soil_row.empty:
             return [{"crop": f"Error: No soil suitability data found for soil type '{soil_type}'", "suitable_for_mixed_cropping": False}]
        # We will use the soil_type provided by the user directly in the rules as soil_df doesn't map to districts


        matches = []
        for _, rule in crop_df.iterrows():
            # Check if the district is in the rule's districts or if "All Districts" is specified
            district_match = False
            if 'districts' in rule and isinstance(rule['districts'], list):
              if district_name in rule["districts"] or "All Districts" in rule["districts"]:
                  district_match = True
            # Check if the season is in the rule's seasons
            season_match = False
            if 'season' in rule and isinstance(rule['season'], list):
              if season in rule["season"]:
                season_match = True
            # Check if the soil is in the rule's soils
            soil_match = False
            if 'soil_type' in rule and isinstance(rule['soil_type'], str):
              if soil_type.lower() in rule['soil_type'].lower():
                soil_match = True
            # If 'soil' column exists and is a list, check if the soil type is in the list
            elif 'soil' in rule and isinstance(rule['soil'], list):
                if soil_type.lower() in [s.lower() for s in rule['soil']]:
                    soil_match = True


            # Check rainfall and groundwater conditions if columns exist and are numeric
            rainfall_match = True
            if 'rainfall_mm' in rule and pd.notna(rule['rainfall_mm']) and pd.notna(rainfall):
              rainfall_match = rule["rainfall_mm"] <= rainfall

            groundwater_match = True
            if 'groundwater_level_m' in rule and pd.notna(rule['groundwater_level_m']) and pd.notna(groundwater):
                # Assuming lower groundwater level (higher value in meters below ground level) is less favorable
                groundwater_match = rule['groundwater_level_m'] >= groundwater # Rule specifies max groundwater level


            temp_match = True # No temperature column in the provided dataframes to use for filtering

            if (
                district_match and
                season_match and
                soil_match and
                rainfall_match and
                groundwater_match and
                temp_match # Placeholder for temperature check
            ):
                matches.append({
                    "crop": rule["crop"],
                    "suitable_for_mixed_cropping": rule.get("mixed_crop", False) # Use .get to handle missing column gracefully
                })

        return matches or [{"crop": "No suitable crop found based on the provided criteria", "suitable_for_mixed_cropping": False}]
    except Exception as e:
        return [{"crop": "Error: " + str(e), "suitable_for_mixed_cropping": False}]

# === 8. Example Usage ===
district = "Guntur"
season = "Rabi"
temperature = 25 # This will not be used in the current recommendation logic as there is no temperature data
soil = "Black" # Add soil type input

recommended = recommend_crops(district, season, temperature, soil)
print(f"\n🌾 Recommended crops for {district} ({season}, Soil={soil}):")
for r in recommended:
    print(f"✅ {r['crop']} | Mixed Cropping: {r['suitable_for_mixed_cropping']}")
```

```python
def recommend_crops(district_name, season, temperature, soil_type):
    try:
        # 1. Environmental Data (rainfall + groundwater)
        env_data = merged_df[merged_df["District"] == district_name]
        if env_data.empty:
            return [{"crop": f"❌ No environmental data found for {district_name}", "suitable_for_mixed_cropping": False}]
        rainfall = env_data.iloc[0]["Rainfall"]
        groundwater = env_data.iloc[0]["Groundwater"]

        # 2. Soil suitability check (basic presence check)
        if soil_df[soil_df["Soil_Type"].str.contains(soil_type, case=False)].empty:
            return [{"crop": f"❌ No soil suitability data for '{soil_type}'", "suitable_for_mixed_cropping": False}]

        # 3. Iterate and apply rule-based filter
        matches = []
        for _, rule in crop_df.iterrows():
            district_match = (
                isinstance(rule.get("districts"), list) and
                (district_name in rule["districts"] or "All Districts" in rule["districts"])
            )
            season_match = (
                isinstance(rule.get("season"), list) and
                season in rule["season"]
            )
            soil_match = (
                isinstance(rule.get("soil"), list) and
                any(soil_type.lower() in s.lower() for s in rule["soil"])
            )
            rainfall_match = (
                pd.notna(rule.get("rainfall_min")) and
                pd.notna(rule.get("rainfall_max")) and
                rule["rainfall_min"] <= rainfall <= rule["rainfall_max"]
            )
            groundwater_match = (
                pd.notna(groundwater)  # since your rules might not have min/max groundwater
            )
            temp_match = (
                pd.notna(rule.get("temp_min")) and
                pd.notna(rule.get("temp_max")) and
                rule["temp_min"] <= temperature <= rule["temp_max"]
            )

            if district_match and season_match and soil_match and rainfall_match and groundwater_match and temp_match:
                matches.append({
                    "crop": rule["crop"],
                    "suitable_for_mixed_cropping": rule.get("mixed_cropping", False)
                })

        return matches or [{"crop": "⚠️ No suitable crop found based on the provided criteria", "suitable_for_mixed_cropping": False}]

    except Exception as e:
        return [{"crop": f"Error: {str(e)}", "suitable_for_mixed_cropping": False}]
```

```python
import pandas as pd
import numpy as np

class CropRecommender:
    def __init__(self):
        """
        Initialize the Crop Recommender system
        """
        self.groundwater_df = None
        self.rainfall_df = None
        self.soil_df = None
        self.crop_df = None
        self.mixed_df = None
        self.rules = {}

    def load_data(self):
        """
        Load and clean all required datasets
        """
        self.groundwater_df = pd.read_csv("/content/drive/MyDrive/GW_Levels_1750139546761.csv")
        self.rainfall_df = pd.read_csv("/content/drive/MyDrive/Rainfall_Summary_1750141935618.csv")
        self.soil_df = pd.read_csv("/content/drive/MyDrive/Soil_Crop_Suitability_AP.csv")
        self.crop_df = pd.read_csv("/content/drive/MyDrive/AP_Crop_Dataset.csv")
        self.mixed_df = pd.read_csv("/content/drive/MyDrive/district_mixed_cropping_suggestions.csv")

        self._preprocess_data()

    def _preprocess_data(self):
        """
        Clean and preprocess all datasets
        """
        dataframes = [self.groundwater_df, self.rainfall_df, self.soil_df, self.crop_df, self.mixed_df]
        for df in dataframes:
            numeric_cols = df.select_dtypes(include=np.number).columns
            df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

        if 'Soil_Type' in self.soil_df.columns:
            self.soil_df['Soil_Type'] = self.soil_df['Soil_Type'].astype('category')

    def define_rules(self):
        """
        Define rule-based crop recommendation rules
        """
        self.rules = {
            'Rice': {
                'soil_type': ['Clay', 'Loamy'],
                'districts': ['East Godavari', 'West Godavari', 'Krishna', 'Guntur', 'Srikakulam'],
                'min_rainfall': 1000,
                'max_rainfall': 1500,
                'min_groundwater': 5,
                'max_groundwater': 15
            },
            'Wheat': {
                'soil_type': ['Loamy', 'Sandy'],
                'districts': ['Anantapur', 'Kurnool'],
                'min_rainfall': 500,
                'max_rainfall': 800,
                'min_groundwater': 3,
                'max_groundwater': 10
            },
            'Maize': {
                'soil_type': ['Sandy', 'Loamy'],
                'districts': ['Prakasam', 'Chittoor', 'Kurnool'],
                'min_rainfall': 600,
                'max_rainfall': 1200,
                'min_groundwater': 4,
                'max_groundwater': 12
            },
            'Chillies': {
                'soil_type': ['Black', 'Sandy Loam'],
                'districts': ['Guntur', 'Prakasam'],
                'min_rainfall': 600,
                'max_rainfall': 900,
                'min_groundwater': 4,
                'max_groundwater': 10
            },
            'Sugarcane': {
                'soil_type': ['Loamy', 'Alluvial'],
                'districts': ['East Godavari', 'West Godavari', 'Krishna'],
                'min_rainfall': 1000,
                'max_rainfall': 1600,
                'min_groundwater': 6,
                'max_groundwater': 15
            },
            'Sunflower': {
                'soil_type': ['Sandy Loam', 'Black'],
                'districts': ['Kurnool', 'Anantapur'],
                'min_rainfall': 400,
                'max_rainfall': 800,
                'min_groundwater': 3,
                'max_groundwater': 10
            },
            'Jowar': {
                'soil_type': ['Black', 'Loamy'],
                'districts': ['Anantapur', 'Kurnool'],
                'min_rainfall': 450,
                'max_rainfall': 800,
                'min_groundwater': 2,
                'max_groundwater': 9
            },
            'Bajra': {
                'soil_type': ['Sandy', 'Loamy'],
                'districts': ['Chittoor', 'Anantapur'],
                'min_rainfall': 300,
                'max_rainfall': 600,
                'min_groundwater': 2,
                'max_groundwater': 8
            },
            'Greengram': {
                'soil_type': ['Sandy Loam', 'Loamy'],
                'districts': ['Kurnool', 'Anantapur'],
                'min_rainfall': 400,
                'max_rainfall': 600,
                'min_groundwater': 3,
                'max_groundwater': 8
            },
            'Bengalgram': {
                'soil_type': ['Black', 'Loamy'],
                'districts': ['Kurnool', 'Kadapa'],
                'min_rainfall': 450,
                'max_rainfall': 700,
                'min_groundwater': 3,
                'max_groundwater': 9
            },
            'Groundnut': {
                'soil_type': ['Red Sandy', 'Loamy'],
                'districts': ['Anantapur', 'Chittoor'],
                'min_rainfall': 500,
                'max_rainfall': 800,
                'min_groundwater': 3,
                'max_groundwater': 9
            },
            'Cotton': {
                'soil_type': ['Black', 'Sandy Loam'],
                'districts': ['Prakasam', 'Kurnool'],
                'min_rainfall': 600,
                'max_rainfall': 1000,
                'min_groundwater': 4,
                'max_groundwater': 12
            },
            'Mango': {
                'soil_type': ['Well-drained loamy', 'Red'],
                'districts': ['Chittoor', 'Vizianagaram'],
                'min_rainfall': 700,
                'max_rainfall': 1200,
                'min_groundwater': 4,
                'max_groundwater': 10
            },
            'Banana': {
                'soil_type': ['Alluvial', 'Loamy'],
                'districts': ['Krishna', 'West Godavari'],
                'min_rainfall': 1000,
                'max_rainfall': 1500,
                'min_groundwater': 5,
                'max_groundwater': 14
            },
            'Corn': {
                'soil_type': ['Loamy', 'Sandy'],
                'districts': ['Prakasam', 'Kurnool'],
                'min_rainfall': 500,
                'max_rainfall': 900,
                'min_groundwater': 3,
                'max_groundwater': 10
            }
        }

    def recommend(self, soil_type: str, rainfall: float, groundwater: float, district: str = None):
        recommendations = {}

        for crop, rules in self.rules.items():
            score = 0

            if soil_type in rules['soil_type']:
                score += 30
            if rules['min_rainfall'] <= rainfall <= rules['max_rainfall']:
                score += 30
            if rules['min_groundwater'] <= groundwater <= rules['max_groundwater']:
                score += 30
            if district and district in rules['districts']:
                score += 10

            if score > 0:
                recommendations[crop] = min(100, score)

        return sorted(recommendations.items(), key=lambda x: x[1], reverse=True)

    def test_recommendation(self):
        test_cases = [
            {'soil_type': 'Clay', 'rainfall': 1200, 'groundwater': 10, 'district': 'Krishna'},
            {'soil_type': 'Loamy', 'rainfall': 600, 'groundwater': 5, 'district': 'Kurnool'},
            {'soil_type': 'Sandy', 'rainfall': 800, 'groundwater': 8, 'district': 'Prakasam'},
            {'soil_type': 'Black', 'rainfall': 1000, 'groundwater': 7, 'district': 'Kadapa'},
        ]

        for i, test_case in enumerate(test_cases):
            print(f"\n🔹 Test Case {i+1}: {test_case}")
            recommendations = self.recommend(**test_case)
            print("✅ Recommendations:")
            for crop, score in recommendations:
                print(f"   - {crop}: {score}% match")

# Main execution
if __name__ == "__main__":
    recommender = CropRecommender()
    recommender.load_data()
    recommender.define_rules()
    print("🌱 Starting crop recommendation tests...")
    recommender.test_recommendation()
    print("\n✅ Recommendation testing completed!")
```
